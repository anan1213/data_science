```python
def kesson_table(df):
    total = df.isnull().sum()
    percent = total / len(df) * 100
    kesson_table = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return kesson_table
```

```python
interval = (-1, 5, 12, 18, 25, 35, 60, 120)
labels = ['baby', 'child', 'teen', 'student', 'young', 'adult', 'senior']

df_train['Age'] = pd.cut(df_train['Age'], interval, labels=labels)
df_test['Age'] = pd.cut(df_test['Age'], interval, labels=labels)
```

```python
df_train.head()
```

```python
Drop = ['Name', 'SibSp', 'Parch', 'Ticket', 'Cabin']
df_train.drop(Drop, axis=1, inplace=True)
df_test.drop(Drop, axis=1, inplace=True)
```

```python
df_train.head()
```

```python
print(df_train.shape, df_test.shape)
```

```python
columns = ['Pclass', 'Sex', 'Age', 'Embarked', 'family']
prefix = ['Pclass', 'Sex', 'Age', 'Embarked', 'family']

df_train = pd.get_dummies(df_train, columns=columns, prefix=prefix, drop_first=True)
df_test = pd.get_dummies(df_test, columns=columns, prefix=prefix, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
train = df_train.drop(['PassengerId', 'Survived'], axis=1)
train_ = df_train['Survived']
test = df_test.drop(['PassengerId'], axis=1)

X_train = sc.fit_transform(train.values)
y_train = train_.values
X_test = sc.fit_transform(test.values)
```

```python
print(X_train.shape, y_train.shape, X_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Dense(20, activation='relu', input_dim=20, kernel_initializer='uniform'))
model.add(Dropout(0.4))
#model.add(Dense(1000, activation='relu', kernel_initializer='uniform'))
#model.add(Dropout(0.4))
model.add(Dense(40, activation='relu', kernel_initializer='uniform'))
model.add(Dropout(0.4))
model.add(Dense(1, activation='sigmoid', kernel_initializer='uniform'))

model.summary()
```

```python
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(X_train, y_train,validation_split=0.2, batch_size=100, epochs=300, verbose=1)

scores = model.evaluate(X_train, y_train, batch_size=30)
print('%s: %.2f%%' % (model.metrics_names[1], scores[1]*100))
```

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoxh')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
predict = np.round(model.predict(X_test))

df_out = pd.read_csv('../data/titanic/test.csv')
df_out['Survived'] = predict
df_out['Survived'] = df_out['Survived'].map({1:0, 0:1})
df_out['Survived'] = df_out['Survived'].map({1:0, 0:1})

df_out[["PassengerId","Survived"]].to_csv("../data/titanic/output/deep_latest.csv",index=False)
```
