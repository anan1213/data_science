```python
import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.layers.recurrent import SimpleRNN
from sklearn.model_selection import train_test_split
```

```python
def sin(x, T=100):
    return np.sin(2.0 * np.pi * x / T)

def toy_problem(T=100, ampl=0.05):
    x = np.arange(0, 2*T+1)
    noise = ampl * np.random.uniform(low=-1.0, high=1.0, size=len(x))
    return sin(x) + noise
```

```python
f = toy_problem()
```

```python
T = 100
length_of_sequence = 2 * T
maxlen = 25

data = []
target = []

for i in range(0, length_of_sequence - maxlen + 1):
    data.append(f[i: i+maxlen])
    target.append(f[i+maxlen])
```

```python
X = np.array(data).reshape(len(data), maxlen, 1)
Y = np.array(target).reshape(len(data), 1)
```

```python
N_train = int(len(data) * 0.9)
N_validation = len(data) - N_train

X_train, X_validation , Y_train, Y_validataion = train_test_split(X, Y, test_size=N_validation)
```

```python
n_in = len(X[0][0])
n_hidden = 20
n_out = len(Y[0])
print("n_in = {}, n_hidden = {}, n_out = {}".format(n_in, n_hidden, n_out))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(SimpleRNN(n_hidden,
                    input_shape=(maxlen, n_in)))
model.add(Dense(n_out))
model.add(Activation('linear'))
```

```python
model.compile(loss='mse', optimizer=Adam())
model.fit(X_train, Y_train, batch_size=10, epochs=500, validation_data=(X_validation, Y_validataion), verbose=1)
```
